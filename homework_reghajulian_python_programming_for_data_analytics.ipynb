```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('titanic/train.csv')
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
passenger_id = df['PassengerId']
survived = df['Survived']
pclass = df['Pclass']
name = df['Name']
sex = df['Sex']
age = df['Age']
sibsp = df['SibSp']
parch = df['Parch']
ticket = df['Ticket']
fare = df['Fare']
cabin = df['Cabin']
embarked = df['Embarked']
```

```python
fig, ax = plt.subplots(1, 5, figsize = (25, 5))
# df['Sex'].value_counts().plot.bar(color = 'blue', ax = ax[0])
sns.countplot('Sex', hue = 'Survived', data = df, ax = ax[0])
sns.countplot('Embarked', hue = 'Survived', data = df, ax = ax[1])
sns.countplot('Pclass', hue = 'Survived', data = df, ax = ax[2])
sns.countplot('SibSp', hue = 'Survived', data = df, ax = ax[3])
sns.countplot('Parch', hue = 'Survived', data = df, ax = ax[4])
# ax.set_ylabel('Population')
# ax.set_title('Sex: Survived vs Dead')
ax[0].set_title('Number of Survivors By Sex')
ax[0].set_ylabel('Passenger')
ax[1].set_title('Number of Survivors By Embarked')
ax[1].set_ylabel('Passenger')
ax[2].set_title('Number of Survivors By Pclass')
ax[2].set_ylabel('Passenger')
ax[3].set_title('Number of Survivors By SibSp')
ax[3].set_ylabel('Passenger')
ax[4].set_title('Number of Survivors By Parch')
ax[4].set_ylabel('Passenger')
plt.show()
```
